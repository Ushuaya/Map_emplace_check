```python
import matplotlib.pyplot as plt
```

```python
with open("results.txt", "r") as file: 
    txt = file.read()
```

```python
txt
```

```python
txt = txt.split('\n')[2:]
```

```python
txt = [i.split(" ") for i in txt]
```

```python
txt = [i[:4] for i in txt]
```

```python
txt = txt[:-1]
```

```python
txt
```

```python
line_insert = [int(i[0]) for i in txt]
```

```python
line_hint = [int(i[1]) for i in txt]
```

```python
line_begin = [int(i[2]) for i in txt]
```

```python
line_random = [int(i[3]) for i in txt]
```

```python
x_num_test = [i + 1 for i in range(len(txt))]
```

```python

  
plt.plot(x_num_test, line_insert, label = "Simple insertion")
plt.plot(x_num_test, line_hint, label = "Insertion to the end")
plt.plot(x_num_test, line_begin, label = "Insertion to the begin")
plt.plot(x_num_test, line_random, label = "Insertion to the random place")
plt.xlabel('Number of test')
plt.ylabel('Time running')

plt.legend()
#plt.show()
plt.savefig("Result.png")
```
